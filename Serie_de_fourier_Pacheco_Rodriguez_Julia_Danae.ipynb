```python
from pylab import *
def f_(x): #se define la funcion a trozos
  y = zeros(len(x))
  for i in range(len(x)):
    if -2<=x[i]<0:
      y[i]=0
    elif 0<=x[i]<=2:
      y[i]=2-x[i]
  return y
def fourier_series(x,N=20): #definimos la serie de fourier
  L=2
  a0=1
  f=a0/2
  for n in range(1,N+1):
    an=(2/(n**2 * pi**2)*(1-cos(n*pi)))
    bn=(2/(n*pi))*(sin(n*pi)-n*pi*cos(n*pi))
    f+=an*cos(n*pi*x/L)+bn*sin(n*pi*x/L)
  return f
#generando los valores para x
x = linspace(-2,2,100)
f_real=f_(x)
#graficando la funcion original
plot(x,f_real,label="funcion original",color="black",linewidth=2)

for N in (2,10):
  plot (x,fourier_series(x,N),label=f'N={N}') #grafica la funcion con diferentes N
xlabel('$x$')
ylabel('$f(x)$')
title("Aproximacion de fourier de la funcion a trozos")
xlim([-2,2]) #configuracion de  de x y y
ylim([-0.5,2.5])
grid();legend()
show()
```
