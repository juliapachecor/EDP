```python
from sympy import fourier_series, pi, plot,exp
from sympy.abc import x
```

```python
f=(exp(2*x))
```

```python
plot(f, (x,0,1),line_color ='red')
```

```python
s = fourier_series(f, (x, 0, 1))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 7)
p = plot(f, s1, s2, s3, (x, -pi, pi), show=False, legend=True)
```

```python
s
```

```python
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
```

```python
p.show()
```

```python
from sympy import fourier_series, pi, plot, Piecewise
from sympy.abc import x
f=Piecewise(
    (1, (0<=x)&(x <=pi/2),
    (2,(pi/2<x)&(x<=pi)),
    ))
plot(f, (x,-1,1),line_color ='red')
s = fourier_series(f, (x, -1, 1))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 7)
p = plot(f, s1, s2, s3, (x, -pi, pi), show=False, legend=True)
s
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
p.show()
```
