```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definir la variable simblica
x, n = sp.symbols('x n')

# Definir la funcin f(x)
f_x = x**2 - 2*x + 3

# Intervalo del perodo
L = 1  # En este caso, el perodo es 2L = 2

# Clculo de coeficientes de Fourier
a_0 = (1 / L) * sp.integrate(f_x, (x, -L, L))
a_n = (1 / L) * sp.integrate(f_x * sp.cos(n * sp.pi * x / L), (x, -L, L))
b_n = (1 / L) * sp.integrate(f_x * sp.sin(n * sp.pi * x / L), (x, -L, L))

# Evaluar coeficientes
a_0_val = float(a_0)

def eval_an_bn(N):
    an_vals = []
    bn_vals = []
    for n in range(1, N + 1):
        an_n = a_n.subs('n', n).evalf()
        bn_n = b_n.subs('n', n).evalf()
        an_vals.append(float(an_n))
        bn_vals.append(float(bn_n))
    return an_vals, bn_vals

# Nmero de trminos en la serie
N = 10
a_n_vals, b_n_vals = eval_an_bn(N)

# Definir la serie de Fourier
def fourier_series(x_val, N):
    sum_series = a_0_val / 2  # Primer trmino
    for n in range(1, N + 1):
        sum_series += a_n_vals[n-1] * np.cos(n * np.pi * x_val / L) + b_n_vals[n-1] * np.sin(n * np.pi * x_val / L)
    return sum_series

# Generar puntos para la grfica
x_vals = np.linspace(-1.5, 1.5, 400)
f_vals = [fourier_series(xi, N) for xi in x_vals]
original_vals = [xi**2 - 2*xi + 3 for xi in x_vals]

# Graficar la funcin original y su serie de Fourier
plt.figure(figsize=(8, 5))
plt.plot(x_vals, original_vals, label="Funcin original $f(x) = x^2 - 2x + 3$", linestyle="dashed")
plt.plot(x_vals, f_vals, label=f"Serie de Fourier con {N} trminos", color="red")
plt.legend()
plt.grid()
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.title("Aproximacin de Fourier de $f(x) = x^2 - 2x + 3$")
plt.show()
```
