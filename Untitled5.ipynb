```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Definir la funcin inicial f(x)
def f(x):
    return x

# Coeficientes de Fourier para la expansin en serie de cosenos
def bn(n):
    integrand = lambda x: f(x) * np.cos(n * np.pi * x)
    result, _ = quad(integrand, 0, 1)
    return 2 * result

# Solucin de la ecuacin de calor
def u(x, t, N=50):
    series_sum = 0
    for n in range(1, N+1):
        Bn = bn(n)
        series_sum += Bn * np.exp(- (n * np.pi) ** 2 * t) * np.cos(n * np.pi * x)
    return series_sum

# Valores de x y t
x_vals = np.linspace(0, 1, 100)
t_vals = [0, 0.01, 0.05, 0.1]

# Graficar la solucin
plt.figure(figsize=(8, 6))
for t in t_vals:
    u_vals = [u(x, t) for x in x_vals]
    plt.plot(x_vals, u_vals, label=f't={t}')

plt.xlabel('x')
plt.ylabel('u(x,t)')
plt.title('Solucin de la ecuacin de calor')
plt.legend()
plt.grid()
plt.show()
```
